```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python

tot_schools = df["school_name"].nunique()
tot_students = df.groupby("school_name")["student_name"].count().sum()

tot_budget = school_data.budget.sum()

avg_math = df["math_score"].mean()
avg_read = df["reading_score"].mean()

math_perc = len(df.loc[df["math_score"] >= 70, :]) / len(df)
read_perc = len(df.loc[df["reading_score"] >= 70, :]) / len(df)

#https://stackoverflow.com/questions/22591174/pandas-multiple-conditions-while-indexing-data-frame-unexpected-behavior
overall_perc = len(df.loc[(df.reading_score >= 70) & (df.math_score >= 70)]) / len(df)

summ1 = pd.DataFrame()
summ1["Total Schools"] = [tot_schools]
summ1["Total Students"] = [tot_students]
summ1["Total Budget"] = [tot_budget]
summ1["Average Math Score"] = [avg_math]
summ1["Average Reading Score"] = [avg_read]
summ1["% Passing Math"] = [math_perc]
summ1["% Passing Reading"] = [read_perc]
summ1["% Overall Passing"] = [overall_perc]

 
```

```python
summ1.style.format({"Total Budget":"${:.2f}", "Average Math Score":"{:.2f}", "Average Reading Score":"{:.2f}", "% Passing Math":"{:.2%}", "% Passing Reading":"{:.2%}", "% Overall Passing":"{:.2%}"})
```

```python
df["passed_math"] = 0

cond = df.math_score >= 70
df.loc[cond, "passed_math"] = 1
math_pass = df.passed_math.sum() / len(df)
math_pass
```

```python
df["passed_reading"] = 0 

cond = df.reading_score >= 70
df.loc[cond, "passed_reading"] = 1
read_pass = df.passed_reading.sum() / len(df)
read_pass
```

```python
df["passed_both"] = 0 

cond = (df.math_score >= 70) & (df.reading_score >= 70)
df.loc[cond, "passed_both"] = 1
both_pass = df.passed_both.sum() / len(df)

both_pass
```

```python
df.head()
```

```python
school_group = df.groupby(["school_name", "type"]).agg({"budget": "mean", "Student ID": "count", "student_name":"count", "reading_score":"mean", "math_score":"mean", "passed_reading":"sum", 
                                                        "passed_math":"sum", "passed_both":"sum"}, axis=1).reset_index().set_index("school_name")

school_group
```

```python
school_group["% Passing Math"] = school_group.passed_math / school_group["Student ID"] * 100
school_group["% Passing Reading"] = school_group.passed_reading / school_group["Student ID"] * 100
school_group["% Passed Both"] = school_group.passed_both / school_group['Student ID'] * 100
school_group["Budget per Student"] = school_group.budget / school_group['Student ID']
school_group.head(1)
```

```python
# https://www.shanelynn.ie/using-pandas-dataframe-creating-editing-viewing-data-in-python/
school_group = school_group.drop(["passed_reading", "passed_math", "passed_both"], axis=1)
school_group
```
