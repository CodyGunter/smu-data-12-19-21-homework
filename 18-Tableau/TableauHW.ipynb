```python
import pandas as pd
import numpy as np
```

```python
months = ["04", "05", "06", "07", "08", "09"]
df_2019 = pd.DataFrame()

for month in months:
    print(month)
    filepath = f"data/2019{month}-citibike-tripdata.csv.zip"
    df = pd.read_csv(filepath, low_memory=False)
    df = df.sample(frac=0.015, random_state=42)
    df_2019 = df_2019.append(df)
    
df_2019 = df_2019.reset_index(drop=True)
#df_2019.info()
```

```python
months = ["04", "05", "06", "07", "08", "09"]
df_2020 = pd.DataFrame()

for month in months:
    print(month)
    filepath2 = f"data/2020{month}-citibike-tripdata.csv"
    df2 = pd.read_csv(filepath2, low_memory=False)
    df2 = df2.sample(frac=0.015, random_state=42)
    df_2020 = df_2020.append(df2)
    
df_2020 = df_2020.reset_index(drop=True)
df_2020.info()
```

```python
df_2020.head()
```

```python
df = pd.concat([df_2019, df_2020], axis=0)
df.info()
```

```python
df = df.dropna()
df.info()
```

```python
mask = df.tripduration < 28800
df = df.loc[mask, :]
df.info()
```
